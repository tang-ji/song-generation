```python
import librosa
import numpy as np 
import pickle
import matplotlib.pyplot as plt
import os
from IPython.display import Audio

from pydub import AudioSegment
```

```python
seperation_path = "output/Hotel_California_Live/"
log = []
```

```python
wav, rate = librosa.load(seperation_path + "vocals.wav")
```

```python
start = 217.8
end = start+2.1
Audio(wav[int(start * rate):int(end * rate)], rate=rate)
```

```python
log.append([start, end, "she calls friends"])
```

```python
log
```

```python
pickle.dump(log, open(seperation_path + "log.pkl", "wb"))
```

```python
def generate_music_list(wav, rate, seperation_path, log, path_name="test/", file_name="file_list.txt"):
    if not os.path.exists(seperation_path + path_name):
        os.makedirs(seperation_path + path_name)
    f = open(seperation_path + path_name + "file_list.txt", "w+")
    for i, l in enumerate(log):
        data = wav[int(float(l[0]) * rate):int(float(l[1]) * rate)]
        librosa.output.write_wav(seperation_path + path_name + str(i) + ".wav", data, rate)
        f.write("data/{}{}.wav|{}|1\n".format(path_name, str(i), str(l[2])))
    f.close()
```

```python
generate_music_list(wav, rate, seperation_path, log, path_name="Hotel_California_Live/", file_name="file_list.txt")
```

## Compose

```python
log = pickle.load(open(seperation_path + "log.pkl", "rb"))
```

```python
wav_new = np.zeros(int((log[-1][1]+3) * rate), dtype=np.float32)
```

```python
for i, l in enumerate(log):
    wav_temp, _ = librosa.load(seperation_path + "gen/{}.wav".format(i))
    wav_new[int(l[0] * rate):int(l[0] * rate)+len(wav_temp)] = wav_temp
```

```python
accompaniment, rate = librosa.load(seperation_path + "accompaniment.wav")
```

```python
com = wav_new + accompaniment[:len(wav_new)]
```

```python
librosa.output.write_wav(seperation_path + "com.wav", com, rate)
```

```python
sound = AudioSegment.from_wav(seperation_path + "com.wav")

sound.export(seperation_path + "com.mp3", format='mp3')
```
