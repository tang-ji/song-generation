```python
def generate_music_list(wav, rate, seperation_path, log, path_name="test/", file_name="file_list.txt"):
    if not os.path.exists(seperation_path + path_name):
        os.makedirs(seperation_path + path_name)
    f = open(seperation_path + path_name + "file_list.txt", "w+")
    for i, l in enumerate(log):
        data = wav[int(float(l[0]) * rate):int(float(l[1]) * rate)]
        librosa.output.write_wav(seperation_path + path_name + str(i) + ".wav", data, rate)
        f.write("data/{}{}.wav|{}|1\n".format(path_name, str(i), str(l[2])))
    f.close()
```

```python
generate_music_list(wav, rate, seperation_path, log, path_name="you_raised_me_up/", file_name="file_list.txt")
```

```python
log = pickle.load(open(seperation_path + "log.pkl", "rb"))
```

```python
wav_new = np.zeros(int((log[-1][1]+3) * rate))
```

```python
for i, l in enumerate(log):
    wav_temp, _ = librosa.load(seperation_path + "gen/{}.wav".format(i))
    wav_new[int(l[0] * rate):int(l[0] * rate)+len(wav_temp)] = wav_temp
```

```python
accompaniment, rate = librosa.load(seperation_path + "accompaniment.wav")
```

```python
com = wav_new + accompaniment[:len(wav_new)]
```

```python
librosa.output.write_wav(seperation_path + "com.wav", com, rate)
```
